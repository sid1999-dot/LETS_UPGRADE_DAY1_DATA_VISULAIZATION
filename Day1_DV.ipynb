```python
x=np.arange(1,11)
y=3*x+5
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine waveform")
plt.plot(x,y,color='r')
plt.show()
```

```python
x=np.arange(0,10)
y=x*x*x
plt.plot(x,y,'ro--')
plt.xlabel('x axis', color='r')
plt.ylabel('y axis', color='b')
plt.title('array drawing')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'ro')
```
